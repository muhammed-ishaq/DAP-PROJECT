```python
import datetime as dt

# Combine 'year' and 'month' to create a datetime column
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# Define today’s date for recency calculation
today_date = dt.datetime(2025, 1, 1)

# Aggregating RFM metrics using 'user_id' as customer identifier
rfm_table = df.groupby('user_id').agg({
    'date': lambda x: (today_date - x.max()).days,  # Recency
    'item_id': 'count',  # Frequency
    'amount': 'sum'  # Monetary
}).rename(columns={'date': 'Recency', 'item_id': 'Frequency', 'amount': 'Monetary'})

# Display top 5 RFM scores
print(rfm_table.head())
```

```python
# Using fixed interval binning with pd.cut for better handling of limited data variation
rfm_table['R'] = pd.cut(rfm_table['Recency'], bins=4, labels=[4, 3, 2, 1])  # Lower recency is better
rfm_table['F'] = pd.cut(rfm_table['Frequency'], bins=4, labels=[1, 2, 3, 4])  # Higher frequency is better
rfm_table['M'] = pd.cut(rfm_table['Monetary'], bins=4, labels=[1, 2, 3, 4])  # Higher monetary value is better

# Combine RFM scores into a single column
rfm_table['RFM_Score'] = rfm_table['R'].astype(str) + rfm_table['F'].astype(str) + rfm_table['M'].astype(str)

# Display top 5 RFM scores
print(rfm_table.head())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing RFM metrics distribution
plt.figure(figsize=(18, 5))

# Distribution of Recency
plt.subplot(1, 3, 1)
sns.histplot(rfm_table['Recency'], kde=True, bins=20, color='blue')
plt.title('Recency Distribution')
plt.xlabel('Recency (Days)')
plt.ylabel('Frequency')

# Distribution of Frequency
plt.subplot(1, 3, 2)
sns.histplot(rfm_table['Frequency'], kde=True, bins=20, color='green')
plt.title('Frequency Distribution')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')

# Distribution of Monetary
plt.subplot(1, 3, 3)
sns.histplot(rfm_table['Monetary'], kde=True, bins=20, color='red')
plt.title('Monetary Distribution')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Ensure RFM scores are categorical if needed (if using them in the heatmap)
rfm_table['R'] = pd.Categorical(rfm_table['R'], categories=[4, 3, 2, 1], ordered=True)
rfm_table['F'] = pd.Categorical(rfm_table['F'], categories=[1, 2, 3, 4], ordered=True)

# Use groupby with observed=True to silence the warning
rfm_score_counts = rfm_table.groupby(['R', 'F'], observed=True).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(rfm_score_counts, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title('RFM Score Heatmap')
plt.xlabel('Frequency Score')
plt.ylabel('Recency Score')
plt.show()
```

```python
# Calculate total sales per transaction
df['total_sales'] = df['amount'] * df['quantity']

# Calculate LTV for each user
ltv = df.groupby('user_id')['total_sales'].sum()

# Average purchase frequency
purchase_frequency = df.groupby('user_id')['item_id'].count().mean()  # Assuming 'item_id' represents individual purchases

# Repeat Rate (number of repeat buyers / unique buyers)
unique_customers = df['user_id'].nunique()
repeat_customers = df[df.groupby('user_id')['item_id'].transform('count') > 1]['user_id'].nunique()
repeat_rate = repeat_customers / unique_customers

# Display LTV and additional metrics
print("Average Purchase Frequency:", purchase_frequency)
print("Repeat Rate:", repeat_rate)
print("Total Customer Lifetime Value (Sample):")
print(ltv.head())
```
